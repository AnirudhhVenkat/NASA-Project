```python
import matplotlib.pyplot as plt

import cartopy.crs as ccrs

import cartopy.feature as cfeature

import cmocean

import cmocean as cmo


def plot_gulf_of_mexico(lon, lat, intensity, lon2, lat2, intensity2):

    # Create a figure and axes with Cartopy projection

    fig = plt.figure(figsize=(20, 20))

    ax = fig.add_subplot(1, 1, 1, projection=ccrs.PlateCarree())

 

    # Set the extent to zoom in even more on the upper part of the Gulf of Mexico region

    ax.set_extent([-150, -60, 10, 60], crs=ccrs.PlateCarree())

 

    # Add coastline, land, and ocean features

    ax.coastlines()

    ax.add_feature(cfeature.LAND, facecolor='lightgray')
    
    ax.add_feature(cfeature.OCEAN)

    #Making colormap red
 
    cmap = plt.cm.get_cmap('Reds')
    
    #second colormap
    
    cmap2 = cmocean.cm.thermal


    # Plot the longitude and latitude coordinates without connecting lines

    sc = ax.scatter(lon, lat, marker='o', c = intensity, cmap=cmap, vmin=129.85, vmax=304.73, s=50, transform=ccrs.PlateCarree())

    sc2 = ax.scatter(lon2, lat2, marker='o', c = intensity2, cmap=cmap2, vmin=21.518, vmax=28.0181, s=150, transform=ccrs.PlateCarree())
    #making colorbar
    
    cb = plt.colorbar(sc, ax=ax, shrink = 0.5, pad = 0.05)
    
    cb.ax.set_title('DOC(umol/kg)')
    
    cb2 = plt.colorbar(sc2, ax=ax, shrink = 0.5, pad = 0.05, location = 'left')
    
    cb2.ax.set_title('SST(mg/m^3)')

    # Set title and display the plot

    ax.set_title('DOC in the Zoomed-in Upper Gulf of Mexico')

    plt.show()
    
    
    
    

 

# Example 

latitude = [28.864425, 28.864425, 28.879088, 28.879088, 28.879088, 28.879088, 28.879088, 28.879088, 28.913445, 28.913445, 28.912474, 29.027052, 29.027052, 29.027052, 29.027052, 29.027052, 29.027052, 28.894333, 28.894333, 28.894333, 29.018283, 29.116283, 29.056633, 28.931933, 28.931933, 28.901817, 28.866233, 28.909150, 29.048033, 29.048033, 28.903050, 28.892233, 28.882390, 28.878650, 28.855450, 28.855450, 28.855450, 28.997500, 28.865500, 28.865500, 28.899567, 28.919217, 28.919217, 28.919217, 28.919217, 28.919217]

longitude = [-89.577282, -89.577282, -89.673118, -89.673118, -89.673118, -89.673118, -89.673118, -89.673118, -89.590828, -89.590828, -89.487556, -89.444388, -89.444388, -89.444388, -89.444388, -89.444388, -89.444388, -89.457333, -89.457333, -89.457333, -89.341200, -89.261000, -89.204183, -89.356467, -89.356467, -89.455200, -89.472900, -89.473300, -89.487033, -89.487033, -89.458483, -89.454667, -89.477023, -89.494417, -89.610950, -89.610950, -89.610950, -89.143525, -89.423450, -89.423450, -89.535133, -89.645600, -89.645600, -89.645600, -89.645600, -89.645600]

latitude2 = [24.50825, 24.44061667, 24.42931667, 24.37575, 24.31421667, 24.28973333, 24.30786667, 24.34533333, 24.40123333, 24.43523333, 24.45431667, 24.4754, 24.5005, 24.549, 24.65453333, 24.76725, 24.88055, 24.97876667, 25.0881, 25.1949, 25.28881667, 25.3845, 25.48415, 25.56443333, 25.65965, 25.80406667, 25.96375, 26.13036667, 26.28916667, 26.45516667, 26.60305, 26.74463333, 26.86903333, 27.01231667, 27.17061667, 27.32491667, 27.46418333, 27.58316667, 27.63851667, 27.67673333, 27.70795, 27.7546, 27.80585, 27.83628333, 27.83778333, 27.81861667, 27.7948, 27.77916667, 27.7529, 27.71515, 27.69323333, 27.71873333, 27.74641667, 27.74318333, 27.69266667, 27.71021667, 27.73541667, 27.72981667, 27.71505, 27.68566667, 27.66295, 27.63828333, 27.61195, 27.59115, 27.57985, 27.58786667, 27.59763333, 27.60743333, 27.6036, 27.59283333, 27.59993333, 27.58315, 27.56006667, 27.49863333, 27.55766667, 27.63995, 27.73521667, 27.8233, 27.84986667, 27.7765, 27.79101667, 27.86435, 27.94335, 28.00521667, 28.05923333, 28.11675, 28.14205, 28.1617, 28.17231667, 28.18855, 28.1974, 28.21053333, 28.22373333, 28.22066667, 28.22035, 28.22305, 28.23826667, 28.24086667, 28.24286667, 28.24415, 28.24186667, 28.23223333, 28.21588333, 28.21691667, 28.22275, 28.23888333, 28.25296667, 28.26905, 28.28198333, 28.29155, 28.2984, 28.30531667, 28.31321667, 28.31813333, 28.32183333, 28.33035, 28.33873333, 28.34575, 28.3556, 28.36818333, 28.38446667, 28.39401667, 28.40118333, 28.40343333, 28.41625, 28.43301667, 28.4464, 28.45343333, 28.45845, 28.4584, 28.45773333, 28.45928333, 28.46021667, 28.4707, 28.47615, 28.47758333, 28.48258333, 28.49181667, 28.50748333, 28.52366667, 28.53761667, 28.54463333, 28.55925, 28.44201667, 28.42166667, 28.39218333, 28.35221667, 28.30316667, 28.25478333, 28.19978333, 28.14608333, 28.1136, 28.08036667, 28.0365, 27.99355, 27.95203333, 27.92451667, 27.95191667, 27.9724, 27.92515, 27.86571667, 27.80521667, 27.74843333, 27.67913333, 27.59443333, 27.51495, 27.43843333, 27.3632, 27.29803333, 27.26646667, 27.25818333, 27.2465, 27.22973333, 27.21118333, 27.1839, 27.15146667, 27.11378333, 27.0661, 27.01035, 26.98738333, 26.97243333, 26.94628333, 26.879, 26.81268333, 26.75668333, 26.70475, 26.6538, 26.5799, 26.46773333, 26.35611667, 26.24575, 26.15226667, 26.13551667, 26.02335, 25.91716667, 25.81201667, 25.70818333, 25.60845, 25.50953333, 25.40395, 25.29635, 25.19155, 25.08536667, 24.98398333, 24.90998333, 24.91231667, 24.82291667, 24.74116667, 24.68896667, 24.65233333, 24.61713333, 24.5946, 24.57235, 24.53903333, 24.50913333, 24.49183333, 24.49428333, 24.51491667, 24.541, 24.536, 24.4787, 24.37626667, 24.27248333, 24.16721667, 24.06978333, 23.97228333, 23.86206667, 23.76243333, 23.6891, 23.71821667, 23.73421667, 23.69623333, 23.66853333, 23.64946667, 23.63963333, 23.63058333, 23.58946667, 23.5595, 23.53005, 23.51665, 23.54726667, 23.47096667, 23.3774, 23.2784, 23.17676667, 23.0742, 22.97536667, 22.87398333, 22.77065, 22.66863333, 22.56003333, 22.46308333, 22.38833333, 22.414, 22.38356667, 22.29411667, 22.20751667, 22.12561667, 22.04088333, 21.9901, 21.94001667, 21.83906667, 21.74176667, 21.64983333, 21.60183333, 21.49025, 21.38508333, 21.28913333, 21.24236667, 21.17546667, 21.06361667, 20.9524, 20.84576667, 20.73473333, 20.62191667, 20.5136, 20.42833333, 20.4334, 20.36411667, 20.26591667, 20.17923333, 20.11251667, 20.02186667, 19.9901, 19.96935, 19.85385, 19.73878333, 19.63091667, 19.52675, 19.42443333, 19.33278333, 19.25368333, 19.24045, 19.23656667, 19.2528, 19.1917, 19.14971667, 19.14008333, 19.15143333, 19.16046667, 19.17736667, 19.186, 19.19333333, 19.1715, 19.16775, 19.12823333, 19.21898333, 19.31411667, 19.38513333, 19.35056667, 19.3545, 19.41905, 19.47591667, 19.53463333, 19.61383333, 19.68848333, 19.76526667, 19.77691667, 19.85555, 19.94068333, 20.0287, 20.11678333, 20.2102, 20.29468333, 20.32018333, 20.30141667, 20.24915, 20.17801667, 20.11443333, 20.0451, 19.98593333, 19.94311667, 19.90655, 19.93948333, 19.98526667, 20.01608333, 19.99725, 20.04541667, 20.08586667, 20.11986667, 20.14055, 20.15445, 20.17045, 20.19105, 20.21073333, 20.24458333, 20.26601667, 20.28803333, 20.31373333, 20.33666667, 20.34561667, 20.36013333, 20.38766667, 20.3902, 20.35855, 20.26053333, 20.21508333, 20.18395, 20.15768333, 20.13613333, 20.116, 20.05418333, 19.96336667, 19.87456667, 19.91153333, 19.97638333, 19.9746, 19.96835, 19.96581667, 19.97503333, 19.98968333, 19.96576667, 19.9561, 19.95503333, 19.95145, 19.94525, 19.93275, 19.90781667, 19.8732, 19.83086667, 19.77943333, 19.72448333, 19.64553333, 19.55611667, 19.54138333, 19.5332, 19.4931, 19.44533333, 19.4039, 19.36711667, 19.33028333, 19.29316667, 19.25603333, 19.20953333, 19.16238333, 19.09281667, 19.02951667, 19.00873333, 18.99418333, 18.95361667, 18.91273333, 18.87811667, 18.85573333, 18.8312, 18.80078333, 18.74383333, 18.70235, 18.66371667, 18.61183333, 18.55416667, 18.4879, 18.41855, 18.3736, 18.36761667, 18.37693333, 18.41326667, 18.4595, 18.45716667, 18.44821667, 18.35661667, 18.2705, 18.19793333, 18.18328333, 18.18286667, 18.21646667, 18.32938333, 18.4361, 18.53786667, 18.53773333, 18.53661667, 18.57066667, 18.61233333, 18.61518333, 18.61945, 18.57931667, 18.50965, 18.4414, 18.44451667, 18.47335, 18.51381667, 18.51076667, 18.48381667, 18.46825, 18.468, 18.46798333, 18.46793333, 18.46783333, 18.4678, 18.46785, 18.46781667, 18.4679, 18.46786667, 18.468, 18.46811667, 18.4684, 18.53223333, 18.58385, 18.61846667, 18.65075, 18.67886667, 18.71528333, 18.73431667, 18.7721, 18.83941667, 18.89811667, 18.93146667, 18.96171667, 18.99421667, 19.03558333, 19.06798333, 19.09908333, 19.13598333, 19.154, 19.18248333, 19.20616667, 19.23041667, 19.26198333, 19.29268333, 19.31721667, 19.34068333, 19.35378333, 19.38116667, 19.42325, 19.46453333, 19.50773333, 19.52515, 19.52538333, 19.47408333, 19.46065, 19.47315, 19.49176667, 19.5286, 19.56715, 19.60873333, 19.64958333, 19.69286667, 19.7533, 19.92623333, 19.98911667, 20.03978333, 20.08515, 20.12071667, 20.1426, 20.11671667, 24.6509, 24.76075, 24.86845, 24.96161667, 25.04398333, 25.11416667, 25.18258333, 25.261, 25.34411667, 25.4142, 25.48273333, 25.54733333, 25.50128333, 25.57893333, 25.66148333, 25.63171667, 25.56011667, 25.46926667, 25.36703333, 25.26685, 25.18908333, 25.25381667, 25.32661667, 25.39868333, 25.47668333, 25.5516, 25.62503333, 25.70145, 25.76861667, 25.81778333, 25.88296667, 25.92911667, 25.88778333, 25.81396667, 25.7039, 25.57066667, 25.43555, 25.30366667, 25.18455, 25.0953, 24.99885, 24.9044, 24.81486667, 24.70928333, 24.58616667, 24.46388333, 24.35033333, 24.25331667, 24.16813333, 24.13453333, 24.06071667, 23.97363333, 23.87471667, 23.94266667, 24.00073333, 24.05555, 24.13193333, 24.20945, 24.27925, 24.30248333, 24.2079, 24.11035, 24.01576667, 23.95958333, 24.06381667, 24.15846667, 24.17718333, 24.20046667, 24.17828333, 24.14591667, 24.17111667, 24.1977, 24.15756667, 24.12256667, 24.1426, 24.13716667, 24.20585, 24.29076667, 24.26636667, 24.28503333, 24.35938333, 24.437, 24.51406667, 24.52343333, 24.52663333, 24.53133333]

longitude2 = [-81.80466667, -81.76876667, -81.67305, -81.6388, -81.63408333, -81.6166, -81.51493333, -81.37861667, -81.23883333, -81.10315, -80.98045, -80.84955, -80.711, -80.57933333, -80.45403333, -80.3232, -80.19611667, -80.09963333, -80.00506667, -79.96103333, -79.90141667, -79.81281667, -79.7326, -79.72431667, -79.7486, -79.79011667, -79.82555, -79.83176667, -79.84201667, -79.87435, -79.8639, -79.8395, -79.82663333, -79.79061667, -79.7478, -79.70381667, -79.664, -79.62513333, -79.58976667, -79.5408, -79.48941667, -79.44603333, -79.38126667, -79.32848333, -79.25841667, -79.20185, -79.13683333, -79.1015, -79.05765, -79.0279, -79.01605, -79.03775, -79.05816667, -79.06985, -79.0559, -79.0555, -79.05948333, -79.06363333, -78.99118333, -78.86801667, -78.74643333, -78.62843333, -78.51105, -78.39788333, -78.28801667, -78.17486667, -78.05593333, -77.93466667, -77.81798333, -77.70111667, -77.6967, -77.70975, -77.72788333, -77.73408333, -77.71865, -77.68811667, -77.65855, -77.63713333, -77.62048333, -77.56156667, -77.5433, -77.49068333, -77.42426667, -77.35591667, -77.28553333, -77.22233333, -77.21405, -77.17831667, -77.09193333, -76.98623333, -76.85911667, -76.7179, -76.57085, -76.43203333, -76.31145, -76.2196, -76.23368333, -76.17215, -76.08128333, -75.9909, -75.8979, -75.80655, -75.7261, -75.6483, -75.55435, -75.56246667, -75.60531667, -75.56381667, -75.49363333, -75.39966667, -75.29345, -75.20096667, -75.10918333, -75.01096667, -74.91028333, -74.81523333, -74.72568333, -74.66705, -74.63776667, -74.65085, -74.67841667, -74.69118333, -74.59451667, -74.50566667, -74.42081667, -74.33463333, -74.2531, -74.16291667, -74.07031667, -73.98406667, -73.89346667, -73.78103333, -73.66803333, -73.69053333, -73.57965, -73.44696667, -73.30308333, -73.1586, -73.00933333, -72.86873333, -72.72388333, -72.5658, -72.45968333, -71.3113, -71.17276667, -71.03736667, -70.89123333, -70.74966667, -70.63788333, -70.52973333, -70.43753333, -70.36411667, -70.26073333, -70.12885, -70.00255, -69.88776667, -69.79015, -69.78076667, -69.76503333, -69.65718333, -69.54255, -69.42855, -69.32351667, -69.208, -69.12946667, -69.05436667, -68.97086667, -68.89955, -68.82293333, -68.78915, -68.75716667, -68.72203333, -68.68965, -68.65111667, -68.59631667, -68.54225, -68.51666667, -68.49565, -68.5102, -68.51545, -68.50135, -68.46511667, -68.43951667, -68.4239, -68.40775, -68.39528333, -68.3817, -68.34926667, -68.27891667, -68.20766667, -68.14485, -68.09238333, -68.0864, -68.02428333, -67.95136667, -67.88048333, -67.81588333, -67.75223333, -67.68881667, -67.62305, -67.55615, -67.47848333, -67.40308333, -67.3342, -67.27565, -67.26368333, -67.18805, -67.1095, -67.06058333, -67.01245, -66.96488333, -66.91963333, -66.8735, -66.83216667, -66.7844, -66.74463333, -66.70546667, -66.65425, -66.61928333, -66.64116667, -66.6776, -66.72995, -66.78578333, -66.84216667, -66.89883333, -66.95316667, -67.00496667, -67.02973333, -67.06435, -67.05603333, -67.06688333, -67.11415, -67.17023333, -67.25673333, -67.33585, -67.376, -67.28706667, -67.19538333, -67.09208333, -67.02638333, -67.04645, -67.0364, -67.06548333, -67.08713333, -67.09963333, -67.11328333, -67.13365, -67.1544, -67.18201667, -67.20493333, -67.22805, -67.26275, -67.27236667, -67.27181667, -67.29731667, -67.30993333, -67.3623, -67.42273333, -67.48385, -67.52308333, -67.56325, -67.62828333, -67.69101667, -67.7532, -67.78216667, -67.84856667, -67.89941667, -67.93258333, -67.94178333, -67.96533333, -68.01881667, -68.0828, -68.14591667, -68.21715, -68.28766667, -68.36363333, -68.43146667, -68.4517, -68.5098, -68.57936667, -68.63816667, -68.6792, -68.75335, -68.77918333, -68.76916667, -68.8202, -68.8781, -68.92513333, -68.9333, -68.93715, -68.94486667, -68.9607, -68.9866, -69.00888333, -69.0298, -69.05921667, -69.1013, -69.12168333, -69.17173333, -69.20985, -69.26998333, -69.31651667, -69.33251667, -69.31073333, -69.22631667, -69.15911667, -69.1444, -69.12788333, -69.2213, -69.31553333, -69.38235, -69.48125, -69.58081667, -69.66666667, -69.74156667, -69.8098, -69.87903333, -69.92845, -70.00561667, -70.08178333, -70.16755, -70.2395, -70.30353333, -70.36536667, -70.38746667, -70.48553333, -70.58303333, -70.68896667, -70.78925, -70.88565, -70.98263333, -71.06151667, -71.11965, -71.22768333, -71.32633333, -71.40795, -71.44683333, -71.52738333, -71.59308333, -71.6275, -71.63838333, -71.66125, -71.69865, -71.74333333, -71.78043333, -71.82433333, -71.86016667, -71.89121667, -71.92806667, -71.95353333, -71.96465, -71.97038333, -72.00786667, -72.015, -72.0373, -72.15305, -72.189, -72.21073333, -72.22261667, -72.23208333, -72.24808333, -72.31131667, -72.39381667, -72.46053333, -72.48956667, -72.5336, -72.52983333, -72.522, -72.51568333, -72.58385, -72.68328333, -72.77861667, -72.8918, -73.0158, -73.13546667, -73.2615, -73.35251667, -73.39563333, -73.45775, -73.53661667, -73.6338, -73.72845, -73.82888333, -73.92001667, -73.93776667, -73.95255, -73.96003333, -74.02926667, -74.12713333, -74.22418333, -74.32868333, -74.43766667, -74.54836667, -74.66436667, -74.77945, -74.86006667, -74.93463333, -74.97466667, -75.0341, -75.14086667, -75.24163333, -75.34028333, -75.42873333, -75.51601667, -75.61211667, -75.68871667, -75.74945, -75.76705, -75.83506667, -75.90526667, -75.9815, -76.05461667, -76.13568333, -76.15746667, -76.19691667, -76.28616667, -76.35015, -76.36828333, -76.39053333, -76.40533333, -76.41521667, -76.4372, -76.45421667, -76.45376667, -76.44853333, -76.48415, -76.52551667, -76.57305, -76.58728333, -76.60546667, -76.67358333, -76.76898333, -76.79348333, -76.81593333, -76.89293333, -76.9959, -77.09976667, -77.11086667, -77.1333, -77.23918333, -77.35465, -77.40646667, -77.41265, -77.41271667, -77.4127, -77.4127, -77.41271667, -77.4127, -77.41271667, -77.41271667, -77.41271667, -77.41271667, -77.41275, -77.4127, -77.41236667, -77.43753333, -77.49161667, -77.56281667, -77.6278, -77.69968333, -77.76805, -77.82716667, -77.89175, -77.95541667, -78.04086667, -78.14203333, -78.24873333, -78.35293333, -78.45576667, -78.57565, -78.7003, -78.82776667, -78.9556, -79.08041667, -79.20291667, -79.3223, -79.43731667, -79.5375, -79.63088333, -79.71326667, -79.74353333, -79.78298333, -79.88188333, -80.02266667, -80.16278333, -80.30028333, -80.40248333, -80.42031667, -80.49553333, -80.61176667, -80.73023333, -80.85455, -80.9778, -81.09916667, -81.22211667, -81.34365, -81.45645, -81.64786667, -81.74743333, -81.84866667, -81.93965, -82.0307, -82.10771667, -82.15613333, -85.48335, -85.46971667, -85.44408333, -85.42103333, -85.3997, -85.38543333, -85.37836667, -85.36481667, -85.34751667, -85.32808333, -85.30886667, -85.28528333, -85.30541667, -85.27451667, -85.22416667, -85.19795, -85.17453333, -85.15543333, -85.15078333, -85.15361667, -85.1686, -85.13513333, -85.09086667, -85.04405, -84.98558333, -84.91801667, -84.84895, -84.78235, -84.72065, -84.68128333, -84.62201667, -84.57316667, -84.5542, -84.54058333, -84.48203333, -84.4099, -84.33488333, -84.25208333, -84.15716667, -84.05855, -83.98508333, -83.92056667, -83.87358333, -83.8804, -83.88341667, -83.8687, -83.84985, -83.82278333, -83.79466667, -83.80138333, -83.7793, -83.7446, -82.4191, -82.3667, -82.32396667, -82.27843333, -82.23163333, -82.19656667, -82.17526667, -82.16761667, -82.13838333, -82.10518333, -82.06365, -82.02501667, -81.98125, -81.9423, -81.96193333, -81.97546667, -81.97481667, -81.96125, -81.97061667, -81.98081667, -81.96613333, -81.9557, -81.952, -81.94438333, -81.89576667, -81.85196667, -81.84916667, -81.83451667, -81.80473333, -81.78668333, -81.80521667, -81.82556667, -81.82908333, -81.8187]

intensity2 = [23.8495, 25.1319, 25.4912, 25.5753, 25.5559, 25.5468, 25.5425, 25.5261, 25.5327, 25.4848, 25.5206, 25.5016, 25.6095, 25.6065, 25.5161, 25.5534, 25.5388, 25.4902, 25.4558, 25.4652, 25.5123, 25.4677, 25.549, 25.5743, 25.6165, 25.6035, 25.5296, 25.5713, 25.6441, 25.5466, 25.5461, 25.5308, 25.5161, 25.5023, 25.55, 25.5179, 25.4484, 25.4351, 25.3827, 25.3677, 25.3338, 25.3023, 25.2796, 25.3292, 25.2984, 25.2661, 25.043, 24.8768, 24.9729, 25.1388, 25.1831, 25.2844, 25.2582, 25.2409, 25.134, 25.0687, 25.0545, 24.9822, 24.9263, 24.4698, 23.7463, 23.6951, 23.3694, 22.7739, 22.8599, 22.9671, 22.6476, 23.3184, 23.534, 23.6562, 23.6689, 23.6855, 23.6859, 23.4344, 23.6162, 23.6464, 23.6057, 23.4932, 23.4156, 23.4762, 23.4444, 23.4012, 23.1119, 23.0885, 22.8497, 22.7627, 22.7728, 22.7551, 22.734, 23.0624, 22.9, 23.3583, 23.2411, 22.971, 22.5359, 22.5663, 22.4735, 22.4781, 22.2949, 22.564, 22.5601, 22.6067, 21.9511, 22.214, 21.9632, 21.9695, 21.9829, 21.9352, 21.9775, 21.997, 21.9604, 21.8579, 21.6709, 21.67, 21.6306, 21.811, 22.1361, 21.583, 21.5658, 21.7421, 21.8728, 22.3654, 21.9099, 21.9901, 21.9475, 21.918, 22.0344, 22.1068, 21.713, 21.7258, 21.6556, 21.7258, 21.6479, 21.603, 21.5325, 21.6713, 21.518, 22.003, 22.2433, 22.2522, 22.2323, 22.2583, 22.1422, 22.6936, 22.7768, 22.7956, 22.8008, 21.9092, 21.9972, 21.9836, 21.7345, 21.5795, 22.6967, 23.0056, 22.976, 23.0503, 23.1405, 23.169, 23.2114, 23.4241, 23.5953, 23.5125, 23.5339, 23.4866, 23.5392, 23.4425, 23.3301, 22.9677, 22.0285, 21.9671, 22.2998, 22.5247, 22.6799, 23.0685, 23.3005, 23.3093, 23.2918, 23.3142, 23.2211, 23.2223, 23.2575, 23.3055, 23.2667, 23.1838, 23.4357, 23.451, 23.485, 23.7256, 23.5239, 23.4156, 23.7385, 23.8309, 23.8129, 23.7099, 23.6555, 23.7752, 23.7936, 23.7563, 23.7799, 23.7684, 23.6605, 23.7311, 23.9651, 24.0417, 24.3937, 24.7897, 24.9452, 25.1138, 25.1412, 25.2837, 25.3699, 25.2302, 25.352, 25.2011, 25.0811, 24.9351, 25.0429, 25.0157, 24.9773, 25.0013, 24.9846, 25.1224, 25.1085, 25.126, 25.1419, 24.9614, 24.8192, 25.0224, 25.2498, 24.8465, 24.9393, 24.6689, 24.87, 24.5887, 24.8281, 25.0026, 25.059, 25.3068, 25.3352, 25.4394, 25.3698, 25.2727, 25.2564, 25.2719, 25.2953, 25.1392, 25.2186, 25.0793, 25.0884, 24.9073, 24.9824, 25.0493, 25.1308, 25.2457, 25.3388, 25.5636, 25.5673, 25.714, 25.6736, 25.482, 25.5805, 25.3361, 25.4057, 25.1623, 25.2965, 25.2353, 25.4611, 25.6723, 25.5936, 25.5615, 25.5956, 25.4028, 25.4218, 25.3745, 25.5127, 25.4261, 25.43, 25.4052, 25.7629, 25.7771, 25.8005, 26.1232, 26.2918, 26.1385, 26.0106, 25.9574, 25.7841, 25.7865, 26.2266, 26.2159, 26.0911, 26.0151, 26.0514, 26.0594, 26.0743, 26.356, 26.464, 26.4225, 26.6458, 26.6049, 26.9979, 26.8983, 27.0724, 26.9428, 27.1962, 26.7765, 26.5973, 26.5751, 26.5278, 25.7451, 25.874, 25.6886, 25.6136, 25.9289, 25.9706, 25.937, 25.9333, 25.9564, 25.9695, 26.0254, 25.9134, 25.91, 25.8616, 25.8676, 25.9383, 26.0123, 25.9643, 25.7596, 25.9765, 25.6771, 25.3735, 25.2613, 25.343, 25.3111, 25.7563, 25.7618, 25.6505, 25.6319, 25.7407, 25.8226, 25.8903, 25.9189, 25.9576, 25.8099, 25.7915, 25.7736, 25.7812, 25.8492, 25.9287, 26.031, 26.1805, 26.3108, 26.6061, 27.0368, 27.1042, 26.9463, 26.4282, 26.6166, 26.507, 26.5225, 26.6129, 26.4002, 26.3236, 26.401, 26.5748, 26.1798, 26.1, 26.101, 26.0496, 26.371, 26.5713, 26.9812, 27.2377, 27.2724, 27.3631, 27.6414, 27.6811, 27.7555, 27.5075, 27.4836, 27.3367, 27.1609, 26.7291, 26.821, 27.1468, 27.1583, 27.1621, 27.1177, 27.1257, 27.081, 26.9772, 26.8949, 26.8751, 26.771, 26.6773, 26.7983, 26.8376, 26.8387, 26.5372, 26.5847, 26.9092, 26.9151, 27.1272, 27.205, 27.2207, 27.2329, 27.1944, 27.1286, 27.1412, 27.1257, 26.9906, 26.9604, 26.9547, 26.9463, 26.9516, 26.9205, 26.8943, 26.8873, 26.9014, 26.9221, 27.0242, 27.0332, 27.2156, 26.9974, 26.6865, 27.0325, 27.0957, 27.2781, 27.2362, 27.2725, 27.0778, 27.0595, 27.0336, 26.9242, 26.9555, 27.0775, 26.7335, 26.7261, 27.1185, 27.21, 27.2238, 27.0696, 26.7789, 26.7976, 26.7569, 26.9186, 27.0228, 27.0978, 27.326, 27.4665, 27.7231, 27.8545, 27.5538, 28.0181, 27.6654, 27.3984, 27.3597, 27.3128, 27.341, 27.3123, 27.3408, 27.3442, 27.3271, 27.406, 27.3791, 27.2831, 27.3143, 27.2804, 27.2799, 27.3328, 27.3499, 27.335, 27.3328, 27.3233, 27.3495, 27.3394, 27.4642, 27.4545, 27.398, 27.2752, 27.2925, 27.3437, 27.2921, 27.1157, 27.0983, 27.1457, 27.0218, 27.0056, 26.9548, 26.8669, 26.8033, 26.6083, 26.6753, 26.6859, 26.7822, 26.8415, 26.9478, 26.9591, 26.9584, 26.962, 26.9165, 26.9011, 26.7874, 26.7806, 26.1042, 26.1108, 26.1436, 26.1308, 26.0929, 26.0159, 26.0585, 26.0673, 26.0884, 26.1233, 26.2226, 26.0338, 26.0223, 25.9835, 26.0565, 26.0759, 26.0651, 26.1664, 26.2556, 26.3387, 26.0031, 26.2994, 26.2329, 26.2322, 26.2028, 26.3196, 26.7127, 26.7238, 26.6275, 26.5895, 26.5841, 26.5186, 26.5075, 26.5169, 26.5374, 26.3797, 26.4319, 25.962, 23.9418, 23.791, 23.9512, 24.4576, 24.7829, 24.6366, 24.326, 24.8196, 24.966, 24.5914, 24.5252, 24.4092, 24.6018, 24.5764, 25.0925, 25.3906, 25.3928, 24.9342, 24.904, 24.8375, 24.903, 24.9273, 24.8451, 24.8917, 25.2013, 25.5359, 25.202, 25.2076, 25.1148, 25.06, 25.1504, 25.1421, 25.0864, 25.0562, 24.9654, 25.2611, 25.2019, 25.296, 25.0963, 25.3095, 25.3981, 25.4483, 25.3562, 25.5164, 25.9102, 25.8929, 25.822, 26.0471]

intensity = [174.67, 211.77, 166.61, 161.62, 172.06, 202.95, 197.98, 209.90, 185.43, 199.01, 203.39, 208.26, 212.17, 219.26, 218.23, 226.21, 224.01

,282.65

,173.72

,228.21

,272.86

,265.38

,267.17

,192.13

,215.48

,280.88

,256.90

,220.55

,178.64

,192.45

,229.36

,293.22

,254.56

,228.29

,184.84

,193.31

,184.63

,304.73

,168.02

,178.57

,206.64

,241.43

,129.85

,168.15

,198.10

,198.91]

plot_gulf_of_mexico(longitude, latitude, intensity, longitude2, latitude2, intensity2)


```
