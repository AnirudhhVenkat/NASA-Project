```python
import pandas as pd
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import cmocean


df = pd.read_csv("C:\\Users\\sripr\\Downloads\\gcoos_2020_01_sea_water_temperature.csv", dtype = float)

df2 = pd.read_csv("C:\\Users\\sripr\\Downloads\\gcoos_2020_12_sea_water_temperature.csv", dtype = float)
low_memory = False

fig = plt.figure(figsize = (20,20))

ax = fig.add_subplot(1, 1, 1, projection=ccrs.PlateCarree())

ax.set_extent([-99, -82.5, 24, 31], crs=ccrs.PlateCarree())

ax.coastlines()

ax.add_feature(cfeature.LAND, facecolor='lightgray')
    
ax.add_feature(cfeature.OCEAN)

gridliner = ax.gridlines(crs=ccrs.PlateCarree(), draw_labels=True)

gridliner.xlabels_top = False

gridliner.xlabels_right = False

gridliner.ylines = False

gridliner.xlines = False

cmap = cmocean.cm.thermal


sc = ax.scatter(df["longitude"], df["latitude"], marker='o', c = df["sea_water_temperature "], cmap=cmap, vmin=9, vmax=36, s=150, transform=ccrs.PlateCarree())


cb = plt.colorbar(sc, ax=ax, shrink = 0.35, pad = 0.05)
    
cb.ax.set_title('Celsius')

ax.set_title("2020_Sea_Water_Temperature_Northern_GOM")

plt.show()
```
